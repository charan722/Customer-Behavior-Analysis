```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.head()
```

```python
## Create a column age_group
labels = ['young_adults','adults','middle_age','senior']
```

```python
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,  # Corrected capitalization
    'Annually': 365,
    'Every 3 Months': 99
}

df['purchases_frequency_days'] = df['frequency_of_purchases'].str.strip().map(frequency_mapping)
```

```python
df[['purchases_frequency_days','frequency_of_purchases']]
```

```python
from sqlalchemy import text

# --- !! SET YOUR NEW DATABASE NAME HERE !! ---
new_db_name = "Project_1"

# This is the SQL command to create the database
# "IF NOT EXISTS" prevents an error if you run it twice
sql_command = text(f"CREATE DATABASE IF NOT EXISTS {new_db_name}")

# We use the 'engine' you already created in the previous step
try:
    with engine.connect() as connection:
        connection.execute(sql_command)
    print(f"Database '{new_db_name}' created successfully (or already exists).")

except Exception as e:
    print(f"An error occurred: {e}")
```
